## Compute an approximation of $pi$  (boostrap method)

- let's drop $n$ random points in a square
- we will define as $inside$ the random points that follow within the area of $1/4$ circle
- $pi$ is then going to be $(4*\sum{inside})/n$

```r
n <- 100000
x <- runif(n); y <- runif(n)   # runif generates random deviates from the Uniform Distribution
# x 
# y
inside <- x^2+y^2 <= 1         # count the dots that fall inside the 1/4 circle
pi <- 4*sum(inside)/n          # approx of pi
# size <- trunc(n/(log(n)))
# size
# idx <- sample.int(trunc(n/(log(n))), replace=FALSE)   # create an index to choose data to plot
idx <- sample.int(n,replace=FALSE) 
# idx
x <- x[idx]; y <- y[idx]; inside <- inside[idx]
# x
# y
# inside
plot(x,y, col=ifelse(inside,"blue","gold"), cex = 0.5, pch = ".", main = sprintf("Bootstrap approx of pi, n=%i, pi=%f", n, pi))
pi
```
